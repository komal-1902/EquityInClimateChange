```python
NER = spacy.load("en_core_web_sm")

def tag_entities(text):
    entities = NER(text)
    region = []
    for word in entities.ents:
        if word.label_ in ("GPE", "LOC"):
            region.append(word.text)
    # displacy.render(entities, style="ent", jupyter=True)
    return set(region)
```
